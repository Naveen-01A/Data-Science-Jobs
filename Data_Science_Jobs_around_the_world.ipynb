EXPLORATORY DATA ANALYSIS

```python
top_10_companies = data['Company'].value_counts()[:10].reset_index()
top_10_companies.columns = ['Company', 'Count']

fig = px.bar(
    data_frame=top_10_companies,
    x='Company',
    y='Count',
    color='Company',
    color_discrete_sequence=px.colors.sequential.turbid,
    text='Count',
    title='Top 10 Companies'
)

fig.update_layout(
    xaxis_title="Companies",
    yaxis_title="Count",
    font=dict(size=13, family="Franklin Gothic")
)

fig.show()
```

Here comes the result of top 10 companies in which Google secures the topmost level, the second topmost secures by the Amazon.com and follows by Accenture, IBM and Microsoft.

```python
top10_roles = data['JobTitle'].value_counts()[:10]
fig = px.bar(y=top10_roles.values,
             x=top10_roles.index,
             color = top10_roles.index,
             color_discrete_sequence=px.colors.sequential.OrRd,
             text=top10_roles.values,
             title= 'Top 10 JobTitle',
             template= 'plotly_dark')
fig.update_layout(
    xaxis_title="JobTitle",
    yaxis_title="count",
    font = dict(size=13,family="Franklin Gothic"))
fig.show()
```

Here is the plot of top 10 JobTitle among which Data Scientist secures the topmost level. The Data Analyst occupied the next most level, then follows by Software Engineer, Data Engineer, Business Analyst in top 5 levels.

```python
my_explode=(0,0,0.1,0.1,0.1)
labels= [x for x in data["Company"].value_counts().sort_values(ascending=False)[0:5].index]
patches,ax, text= plt.pie(data["Company"].value_counts().sort_values(ascending=False)[0:5],autopct='%1.1f%%',explode=my_explode,shadow=True,startangle=305)
plt.title("Top 5 Companies with Most Number of Data Science Related Jobs", size=15, color='black')
plt.legend(patches, labels, loc=4, bbox_to_anchor = (1.3,0.25))
plt.axis('equal')
plt.show()

```
